```python
import seaborn as sns
from pandas.api.types import CategoricalDtype
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
cut_categoriler = ["Fair", "Good", "Very Good", "Premium", "Ideal"]
df.cut = df.cut.astype(CategoricalDtype( categories=cut_categoriler,ordered=True))
df.head()
```

```python
sns.catplot(x="cut", y="price", data=df); # .catplot() kategorik veriler için kullanılır. x ve y parametresi x ve y eksenlerini belirler. data parametresi ise veri setini belirler.
```

```python
sns.barplot(x="cut", y="price", hue="color", data=df); # barplot() ise bar grafiği oluşturur. hue parametresi ise renk kodlaması yapar.
```

```python
sns.displot(df.price, kde=True); # displot() dağılım grafiği oluşturur. kde parametresi ise yoğunluk grafiği oluşturur. df.price parametresi ise fiyat sütununu belirtir.
```

```python
sns.displot(df.price, bins = 1000, kde=True); # bins parametresi ise grafiğin çözünürlüğünü belirler.
```

```python
sns.displot(df.price); 
```

```python
sns.kdeplot(df.price, shade=True); # kdeplot() yoğunluk grafiği oluşturur. Shade parametresi ise altını doldurur.
```

```python
(sns.FacetGrid(df, hue="cut", height=5, xlim=(0, 5000)).map(sns.kdeplot, "price", shade=True).add_legend()); 
"""
FacetGrid() ise alt grafik oluşturur. 
hue parametresi ise renk kodlaması yapar. 
height parametresi ise grafik boyutunu belirler. xlim parametresi ise x ekseninin sınırlarını belirler. 
map() fonksiyonu ise grafik türünü belirler. add_legend() ise renk kodlamasını gösterir."""
```

```python
sns.catplot(x="cut", y="price", hue="color", kind="point", data=df); # kind parametresi ise grafik türünü belirler.
```
