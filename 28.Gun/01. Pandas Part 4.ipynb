# Gözlem ve Değişken Seçimi

```python
import numpy as np
import pandas as pd
m = np.random.randint(1,30, size=(10,3))
df = pd.DataFrame(m, columns=["var1","var2","var3"])
df
```

### loc
- Tanımlandığı şekliyle seçim yapmak için kullanılır.

```python
df.loc[0:3]
```

### iloc
- Alışık olduğumuz indexleme mantığıyla seçim yapmak için kullanılır.

```python
df.iloc[0:3]
```

```python
df.iloc[0,0]
```

```python
df.iloc[:3,:2] # 0,1,2 satırların 0,1 sütunları alınır
```

```python
df.loc[0:3,"var3"] # 0,1,2,3 satırların var3 sütunu alınır
```

```python
df.iloc[0:3,0:2] # 0,1,2 satırların 0,1 sütunları alınır
```

## Koşullu Eleman İşlemleri

```python
df[df.var1 > 15] # var1 sütunundaki değerler 15 den büyük olanları getirir
```

```python
df[df.var1 % 2 == 0] # var1 sütunundaki değerler 2 ye bölümünden kalanı 0 olanları getirir
```

```python
df[(df.var1 > 15) & (df.var3 < 15)] # var1 sütunundaki değerler 15 den büyük var3 15 den küçük olanları getirir
```

```python
df.loc[(df.var1 > 15) & (df.var3 < 15), "var2"] # var1 sütunundaki değerler 15 den büyük var3 15 den küçük olanları getirir
```

```python
df.query("var1 > 15 & var3 < 15") # var1 sütunundaki değerler 15 den büyük var3 15 den küçük olanları getirir
```

## Birleştirme (Join) İşlemleri

```python
df2 = df + 99
df2
```

**.concat()** fonksiyonu ile iki dataframe'i birleştirebiliriz. ignore_index parametresi ile indexleri sıfırlayabiliriz.

join parametresi ile hangi kolon üzerinden birleştirileceğini belirleyebiliriz.

```python
pd.concat([df,df2])
```

```python
pd.concat([df,df2], ignore_index=True)
```

```python
df2.columns = ["var1","var2","deg3"]
```

```python
pd.concat([df,df2], join="inner")
```
